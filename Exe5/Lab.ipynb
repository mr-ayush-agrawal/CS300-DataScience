```python
class setcls:
    elements = []
    def __init__(self):
        pass

    def add(self,num):
        if num not in self.elements :
            self.elements.append(num)
    def remove(self,ele):
        if ele in self.elements:
            self.elements.remove(ele)
    def printSet(self): 
        for x in self.elements:
            print (x, end= "\t")
        print()
    def isPresent(self, ele):
        if ele in self.elements:
            return True
        else:
            return False
        

if __name__ == '__main__' :
    s1 = setcls()
    s1.printSet()
    s1.add(12)
    s1.add(1)
    s1.add(2)
    s1.add(132)
    s1.printSet()
    s1.remove(132)
    s1.remove(5)
    s1.printSet()
    print(s1.isPresent(23))
    print(s1.isPresent(12))
```

```python
import datetime
class bank :
    Bal = 0
    def __init__(self):
        pass
    def deposit(self, Amt):
        if Amt <= 0: 
            print("Can't deposit the amount")
            return Bal
        Bal+=Amt
        with open('st.txt','a') as s :
            s.write(f"{datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')} \t\t +{Amt}\t\t {Bal}\n")

    def withdraw(self, Amt):
        if Amt <= 0: 
            print("Can't Withdraw the amount")
            return Bal
        Bal-=Amt
        with open('st.txt','a') as s :
            s.write(f"{datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')} \t\t +{Amt}\t\t {Bal}\n")
    
    def CheckBal(self):
        return Bal
    
    def Printst(self):
        with open('st.txt','r') as s :
            print(s.read())


Ac = bank()
print(Ac.CheckBal())
Ac.deposit(100)
Ac.deposit(1200)
Ac.deposit(1500)
print(Ac.CheckBal())
Ac.withdraw(500)
Ac.Printst()
```

```python
class stk :
    def __init__(self):
        self.elements= []

    def push(self, ele):
        self.elements.append(ele)
        print("Element added to the stack")

    def pop (self):
        if len(self.elements> 0):
            self.elements.pop()

    def peek(self):
        if len(self.elements> 0):
            return self.elements[-1]

    def printstk(self):
        print(self.elements)    
```
