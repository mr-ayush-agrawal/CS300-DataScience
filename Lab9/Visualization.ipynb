```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

pop = pd.read_csv('pop.csv',index_col=0)
tab = pop.pivot_table(columns='country', index='year', values='population')
tab.head()

plt.plot(tab['China'], c='r',label = 'China')
plt.plot(tab['United States'], c='b',label = 'United States')
plt.plot(tab['India'], c='orange',label = 'India')
plt.xlabel("Years")
plt.ylabel("Population in Billions")
plt.legend()
```
