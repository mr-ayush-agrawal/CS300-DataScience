```python
lifeYear = data.groupby(['continent','year'])['lifeExp'].mean().reset_index()
lifeYear
```

```python
piv = lifeYear.pivot_table(index='continent',columns='year',values='lifeExp')
piv
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(piv,cmap='Greens',annot= True)
```
