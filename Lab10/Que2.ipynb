```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Ecommerce.csv')
data.head()
# data.columns
```

```python
data.describe()
```

```python
cols = ['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']
kind = ['scatter', 'hex', 'kde', 'reg']
color = ['r', 'b', 'orange', 'g']
plt.figure(figsize=(6,6))
for x in range(len(cols)):
    sns.jointplot(x = cols[x], y='Yearly Amount Spent', kind = kind[x], data=data, color=color[x])
```
