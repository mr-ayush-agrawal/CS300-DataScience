```python
encoded = pd.get_dummies(df,columns=['Fuel_Type', 'Seller_Type', 'Transmission'])
encoded
```

```python
encoded.info()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(encoded['Present_Price'], encoded['Selling_Price'], alpha=.6, c='r')
plt.xlabel('Present_Price')
plt.ylabel('Selling_Price')
plt.title("Present Price vs Selling Price")
```

```python
encoded.head()
```

```python
x= encoded.drop(columns=['Car_Name', 'Selling_Price'])
y=encoded['Selling_Price']
print(type(x))
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
stdsclr = StandardScaler()
x = stdsclr.fit_transform(x)
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
print(type(x_test))
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

y_test_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MSE = mean_squared_error(y_test, y_test_pred)
MAE = mean_absolute_error(y_test, y_test_pred)
R2 = r2_score(y_test, y_test_pred)

print("Mean SQ Error in the Model Predeciton = ", MSE)
print("Mean Abs Error in the Model Predeciton = ", MAE)
print("R2 Score in the Model Predeciton = ", R2)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_test_pred, c='#212121', alpha=.8)
plt.xlabel("Y Test")
plt.ylabel("Y Test Predectied")
plt.show()
```
