# PreProcessing

```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
data = pd.read_csv('SemiProcessed.csv',index_col=0)
data.head()
```

```python
data['Day'] = data['Day'].astype('int8')
data['Month'] = data['Month'].astype('int8')
data['Year'] = data['Year'].astype('int16')

data = data.rename(columns={'Weekday' : 'day_index'})
data['day_index'] = data['day_index'].astype('int8')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Breakfast_Wastage'] = data['Breakfast_Wastage']/1.5
data['Lunch_Wastage'] = data['Lunch_Wastage']/1.7
data['Snacks_Wastage'] = data['Snacks_Wastage']/2.7
data['Dinner_Wastage'] = data['Dinner_Wastage']/2

data['Breakfast_Wastage'] = data['Breakfast_Wastage'].apply(lambda x : round(x,2))
data['Lunch_Wastage'] = data['Lunch_Wastage'].apply(lambda x : round(x,2))
data['Snacks_Wastage'] = data['Snacks_Wastage'].apply(lambda x : round(x,2))
data['Dinner_Wastage'] = data['Dinner_Wastage'].apply(lambda x : round(x,2))

data['Total_Daily_Wastage'] = data['Breakfast_Wastage'] + data['Lunch_Wastage'] + data['Snacks_Wastage'] + data['Dinner_Wastage']

data.describe()
```
